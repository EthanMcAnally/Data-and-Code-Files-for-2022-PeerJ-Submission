```python
#import data from FASTQ pass folder on MinION device. Only files of used barcodes were retained, then sorted into folders by primer
import os
import glob

WB_S_locations = r"/Users/mcanally23/Documents/WB_fastq_pass/18S"
WB_R_locations = r"/Users/mcanally23/Documents/WB_fastq_pass/rbcL"
FC_S_locations = r"/Users/mcanally23/Documents/FC_fastq_pass/18S"
FC_R_locations = r"/Users/mcanally23/Documents/FC_fastq_pass/rbcL"
```

```python
#creates new concatenated files (one for each barcode) of all files in a barcode folder
os.chdir(WB_S_locations)

for file in WB_S_list:
    wildcard = file[:-12] + '*' + '.fastq'
    final_name = file[:-13] + '.fastq'
    !cat $wildcard > $final_name

os.chdir(WB_R_locations)

for file in WB_R_list:
    wildcard = file[:-12] + '*' + '.fastq'
    final_name = file[:-13] + '.fastq'
    !cat $wildcard > $final_name
    
os.chdir(FC_S_locations)

for file in FC_S_list:
    wildcard = file[:-12] + '*' + '.fastq'
    final_name = file[:-13] + '.fastq'
    !cat $wildcard > $final_name
    
os.chdir(FC_R_locations)

for file in FC_R_list:
    wildcard = file[:-12] + '*' + '.fastq'
    final_name = file[:-13] + '.fastq'
    !cat $wildcard > $final_name
```

```python
#removes old, unconcatenated files
old_WB_S_files = glob.glob(os.path.join(WB_S_locations, '*_filt.fastq'))
old_WB_R_files = glob.glob(os.path.join(WB_R_locations, '*_filt.fastq'))
old_FC_S_files = glob.glob(os.path.join(FC_S_locations, '*_filt.fastq'))
old_FC_R_files = glob.glob(os.path.join(FC_R_locations, '*_filt.fastq'))
for file in old_WB_S_files:
    os.remove(file)
for file in old_WB_R_files:
    os.remove(file)
for file in old_FC_S_files:
    os.remove(file)
for file in old_FC_R_files:
    os.remove(file)
```

```python
##runs cutadapt targeting forward and reverse primers, for 18S and rbcL, for data from both flow cell runs

#runs cutadapt on 18S WB reads for forward adapter (for 18S-528F)
os.chdir(WB_S_locations) # go to the directory of where files so when run command line saves them there 
for file in WB_S_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_18S_F.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g GCGGTAATTCCAGCTCCAA...GTAGGTGAACCTGCAGAAGGATCA --discard-untrimmed -o $new_file_name $file

#runs cutadapt on 18S WB reads for reverse adapter (for 18S-1528R)
for file in WB_S_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_18S_R.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g TGATCCTTCTGCAGGTTCACCTAC...TTGGAGCTGGAATTACCGC --discard-untrimmed -o $new_file_name $file

#runs cutadapt on 18S FC reads for forward adapter (for 18S-528F)
os.chdir(FC_S_locations) # go to the directory of where files so when run command line saves them there 
for file in FC_S_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_18S_F.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g GCGGTAATTCCAGCTCCAA...GTAGGTGAACCTGCAGAAGGATCA --discard-untrimmed -o $new_file_name $file
    
#runs cutadapt on 18S FC reads for reverse adapter (for 18S-1528R)
for file in FC_S_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_18S_R.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g TGATCCTTCTGCAGGTTCACCTAC...TTGGAGCTGGAATTACCGC --discard-untrimmed -o $new_file_name $file
    
os.chdir(WB_R_locations)
#runs cutadapt on rbcL WB reads for forward adapter
for file in WB_R_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_rbcL_F.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g AAGGAGAAATHAATGTCT...GAGACTTACACAAGGTTGYTT --discard-untrimmed -o $new_file_name $file

#runs cutadapt on rbcL WB reads for reverse adapter
for file in WB_R_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_rbcL_R.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g AARCAACCTTGTGTAAGTCTC...AGACATTDATTTCTCCTT --discard-untrimmed -o $new_file_name $file

os.chdir(FC_R_locations)
#runs cutadapt on rbcL FC reads for forward adapter
for file in FC_R_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_rbcL_F.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g AAGGAGAAATHAATGTCT...GAGACTTACACAAGGTTGYTT --discard-untrimmed -o $new_file_name $file

#runs cutadapt on rbcL FC reads for reverse adapter
for file in FC_R_list: # run through each file in the filelist
    new_file_name = file[:-13] + '_rbcL_R.fastq' # includes pathway
    # run command line prompt using file name and new file name
    !cutadapt -g AARCAACCTTGTGTAAGTCTC...AGACATTDATTTCTCCTT --discard-untrimmed -o $new_file_name $file
    
    
##old files were deleted manually in Finder
```
